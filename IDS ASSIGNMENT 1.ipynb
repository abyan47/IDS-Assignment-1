<h4>Q1. Arithmetic and Assignment Operators</h4>

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)  # Returns an integer
print("Modulus:", a % b)           # Returns the remainder
print("Exponentiation:", a ** b)

a += 5
print("a +=5 =",a )

a -= 5
print("a -=5 =",a )

a *= 5
print("a *=5 =",a )

a /= 5
print("a /=5 =",a )
```

<h4>Q2. Comparison and Logical Operators</h4>

```python
x = int(input("Enter 1st number: "))
y = int(input("Enter 2nd number: "))

print(f"{x} == {y}:", x == y)
print(f"{x} != {y}:", x != y)
print(f"{x} > {y}:", x > y)
print(f"{x} < {y}:", x < y)
print(f"{x} <= {y}:", x <= y)
print(f"{x} >= {y}:", x >= y)

print(f"{x} > 0 and {y} > 0:", x>0 and y>0)
print(f"{x} > 0 or {y} > 0:", x>0 or y>0)
print(f"not ({x} > 0):", not (x>0)  )

```

<h4>Q3. Bitwise Operators</h4>

```python
int_1 = 10
int_2 = 20
print("AND:", int_1 & int_2 )
print("OR:", int_1 | int_2 )
print("XOR:", int_1 ^ int_2 )
print("NOT int_1:", ~int_1 )
print("Left Shift int_1:", int_1<<1 )
print("Right Shift int_1:", ~int_1>>1 )
```

<h4>Q4. Identity and Membership Operators</h4>

```python
num = [1,2,3,4,5]
new_num =  [1,2,3,4,5]
copy_num = num

print("num is copy num:", num is copy_num)
print("num is copy new_num:", num is new_num)
print("num is copy num:", num is not new_num)


print('3 in copy num:', 3 in num)
print('3 in copy num:', 3 not in num)
```

<h4>Q5. String Operations</h4>

```python
abyan = "My name is Abyan Khan"
print('Lenth of string:', len(abyan))
print('Upper case of string:', abyan.upper())
print('lower case of string:', abyan.lower())
print('A in string:', 'Abyan Khan' in abyan)
new_abyan =abyan.replace("Abyan","Rohan")
print("Replace in string:", new_abyan)
print("Slice:", abyan[10:16])
print("Reverse", abyan[::-1])
```
